```python
# Функция для расчёта коэффициента
def calculate_coefficient(zone1, zone2):
    common = len(zones[zone1] & zones[zone2])
    total = max(len(zones[zone1]), len(zones[zone2]))
    return common / total

# Создание матрицы совместимости
zone_names = list(zones.keys())
matrix = pd.DataFrame(index=zone_names, columns=zone_names)

for z1, z2 in product(zone_names, repeat=2):
    matrix.loc[z1, z2] = calculate_coefficient(z1, z2)
```

```python
def matrix_to_dict(matrix):
    """
    Преобразует матрицу совместимости в словарь.

    Args:
        matrix (pd.DataFrame): Матрица совместимости.

    Returns:
        dict: Словарь, представляющий данные из матрицы.
    """
    transition_weights = {}
    for z1 in matrix.index:
        transition_weights[z1] = {}
        for z2 in matrix.columns:
            transition_weights[z1][z2] = matrix.loc[z1, z2]
    return transition_weights

# Пример использования
result_dict = matrix_to_dict(matrix)
print(result_dict)
```

```python
def calculate_zone_transition_probabilities(blocks_gdf, transition_weights):
    """
    Рассчитывает вероятности переходов между зонами для каждого квартала.
    
    :param blocks_gdf: GeoDataFrame с кварталами. Должен содержать столбец 'ZONING_tag'.
    :param transition_weights: Словарь весов переходов между зонами. Например:
        {
            "RESIDENTIAL": {"RESIDENTIAL": 1, "RECREATIONAL": 0.35, "PUBLIC_AND_BUSINESS": 0.52},
            "RECREATIONAL": {"RESIDENTIAL": 0.35, "RECREATIONAL": 1, "PUBLIC_AND_BUSINESS": 0.5},
            ...
        }
    :return: GeoDataFrame с добавленным столбцом 'transition_probabilities'.
    """
    # Проверяем наличие столбца ZONE
    if 'ZONING_tag' not in blocks_gdf.columns:
        raise ValueError("GeoDataFrame must contain a 'ZONING_tag' column.")

    # Создаем столбец для хранения словаря переходов
    blocks_gdf['transition_probabilities'] = None

    # Поиск соседей через spatial join
    blocks_with_neighbors = gpd.sjoin(blocks_gdf, blocks_gdf, predicate='intersects', how='left', rsuffix='_neighbor')

    # Итерируемся по каждому кварталу
    for index, block in blocks_gdf.iterrows():
        current_zone = block['ZONING_tag']  # Текущая зона квартала
        neighbors = blocks_with_neighbors[blocks_with_neighbors.index == index]

        # Получаем соседние зоны
        neighbor_zones = neighbors['ZONING_neighbor'].value_counts().to_dict()

        # Суммируем веса для соседних зон
        total_weight = 0
        zone_weights = {}
        for neighbor_zone, count in neighbor_zones.items():
            weight = transition_weights.get(current_zone, {}).get(neighbor_zone, 0)
            zone_weights[neighbor_zone] = count * weight
            total_weight += count * weight

        # Рассчитываем вероятности перехода
        transition_probabilities = {zone: weight / total_weight for zone, weight in zone_weights.items() if total_weight > 0}

        # Сохраняем результат
        blocks_gdf.at[index, 'transition_probabilities'] = transition_probabilities

    return blocks_gdf
```
