```python
# Преобразуем значения в relation_gdf['tags']
def tags_to_dict(tag_list):
    result = {}
    for tag in tag_list:
        if isinstance(tag, Tag):  # если уже объект Enum
            result[tag.name] = tag.value
        elif isinstance(tag, str):  # если просто строка, как "Tag.WATER_OBJECT"
            tag_name = tag.split('.')[-1]
            try:
                tag_enum = Tag[tag_name]
                result[tag_enum.name] = tag_enum.value
            except KeyError:
                result[tag_name] = "Неизвестный тег"
    return result

# Применим функцию к столбцу 'tags'
relation_gdf['tags_dict'] = relation_gdf['tags'].apply(tags_to_dict)
```

```python
def aggregate_requirements_for_tags(tag_list):
    # Преобразуем к списку Tag, если строка — пропускаем
    if not tag_list or not isinstance(tag_list, list):
        return pd.Series({req: None for req in REQUIREMENTS_MATRIX.columns})

    # Получаем подматрицу по тегам
    try:
        sub_matrix = REQUIREMENTS_MATRIX.loc[tag_list]
    except KeyError:
        print(f"Некоторые теги не найдены в матрице: {tag_list}")
        return pd.Series({req: None for req in REQUIREMENTS_MATRIX.columns})

    result = {}
    for requirement in REQUIREMENTS_MATRIX.columns:
        values = sub_matrix[requirement].dropna().unique()
        if False in values:
            result[requirement] = False
        elif True in values:
            result[requirement] = True
        else:
            result[requirement] = None
    return pd.Series(result)

# Применяем к каждому кварталу
requirement_columns = pd.DataFrame(relation_gdf['tags'].apply(aggregate_requirements_for_tags))

# Объединяем с relation_gdf
relation_gdf = pd.concat([relation_gdf, requirement_columns], axis=1)
```

```python
# Добавляем словарь требований
relation_gdf['req_dict'] = relation_gdf[REQUIREMENTS_MATRIX.columns].to_dict(orient='records')

# Оставляем только два столбца
relation_gdf = relation_gdf[['tags_dict', 'req_dict']]
```

```python
from my_package.models.tag import Tag
from my_package.models.requirement import Requirement

# Словарь, где каждому требованию соответствует тег, наличие которого его удовлетворяет
REQUIREMENT_TO_TAG = {
    Requirement.TRANSPORT_PLACEMENT_DISTRICT: Tag.DISTRICT_LEVEL,
    Requirement.TRANSPORT_PLACEMENT_LOCAL: Tag.LOCAL_LEVEL,
    
    Requirement.COMMERCIAL_PLACEMENT: Tag.COMMERCIAL_FACILITIES,
    
    Requirement.PUBLIC_SPACE_CITY: Tag.CITY_EMBANKMENT,
    Requirement.PUBLIC_SPACE_DISTRICT: Tag.CITY_EMBANKMENT,
    #Requirement.PUBLIC_SPACE_LOCAL: Tag.LOCAL_LEVEL,
    
    Requirement.PROTECTION_ZONE: Tag.CITY_FOREST,  # пример, можно поменять
    
    Requirement.OBJECT_PLACEMENT_MEDICAL: Tag.MEDICAL_FACILITIES,
    Requirement.OBJECT_PLACEMENT_EDUCATIONAL: Tag.EDUCATIONAL_FACILITIES,
    Requirement.OBJECT_PLACEMENT_SPORTS: Tag.SPORTS_FACILITIES,
    Requirement.OBJECT_PLACEMENT_CULTURAL: Tag.CULTURAL_FACILITIES,
    Requirement.OBJECT_PLACEMENT_GREENERY: Tag.LINEAR_GREENERY,
    Requirement.OBJECT_PLACEMENT_LANDMARK: Tag.CITY_LANDMARK,

    Requirement.HEIGHT_LOW_RISE: Tag.LOW_RISE,
    Requirement.HEIGHT_MID_RISE: Tag.MID_RISE,
    Requirement.HEIGHT_HIGH_RISE: Tag.HIGH_RISE,
}
```

```python
def separate_requirements(row):
    tags_dict = row['tags_dict']
    
    # Очищаем req_dict от значений None
    req_dict = {k: v for k, v in row['req_dict'].items() if v is not None}
    
    sec_req = {}
    first_req = {}

    for req, value in req_dict.items():
        if value is not True:
            first_req[req] = value
            continue

        # Получаем тег, который должен быть в tags_dict
        tag = REQUIREMENT_TO_TAG.get(req, None)

        if tag is not None and tag.name in tags_dict:
            # Если нужный тег есть — требование выполнено
            sec_req[req] = value
        else:
            first_req[req] = value

    return pd.Series({'first_req': first_req, 'sec_req': sec_req})
```

```python
relation_gdf[['first_req', 'sec_req']] = relation_gdf.apply(separate_requirements, axis=1)
```
